```python
# import das bibliotecas
#acesso ao data frame, neste caso estou usando vscode
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv('Stores.csv',sep=',')
```

```python
print('Items_Available')
itens_disponiveis = df['Items_Available']
max = itens_disponiveis.values.max()
min = itens_disponiveis.values.min()
mean = itens_disponiveis.values.mean()
std = itens_disponiveis.values.std()
print(f'Valor Máximo: {itens_disponiveis.values.max():.2f}')
print(f'Valor Mínimo: {itens_disponiveis.values.min():.2f}')
print(f'Média: {itens_disponiveis.values.mean():.2f}')
print(f'Desvio Padrão: {itens_disponiveis.values.std():.2f}')

#criação do gráfico em barras
labels = ['Max','Min','Média','Desvio Padrão']
data = [max,min,mean,std]
plt.title('Itens Disponiveis')
plt.bar(labels,data)
plt.show()

```

```python
print('Daily_Customer_Count')
contador_cliente = df['Daily_Customer_Count']
max = contador_cliente.values.max()
min = contador_cliente.values.min()
mean = contador_cliente.values.mean()
std = contador_cliente.values.std()
print(f'Valor Máximo: {contador_cliente.values.max():.2f}')
print(f'Valor Mínimo: {contador_cliente.values.min():.2f}')
print(f'Média: {contador_cliente.values.mean():.2f}')
print(f'Desvio Padrão: {contador_cliente.values.std():.2f}')


#criação do gráfico em barras
labels = ['Max','Min','Média','Desvio Padrão']
data = [max,min,mean,std]
plt.title('Contador De Fluxo Diário')
plt.bar(labels,data)
plt.show()
```
